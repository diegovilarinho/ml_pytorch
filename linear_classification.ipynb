## Linear Classification
Generate a binary classification dataset

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46)

X, Y = make_classification(
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_clusters_per_class=1,
)
```

```python
def plotModel(w1, w2, b):
    x = np.linspace(-2, 4, 50)
    y = -w1/w2 * x - b/w2
    
    plt.scatter(X[:, 0], X[:, 1], c=Y, marker='o', edgecolors='k')
    
    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()

    plt.axvline(0, -1, 1, color='black', lw=0.9)
    plt.axhline(0, -2, 1, color='black', lw=0.9)
    plt.plot(x, y)
    plt.grid(True)
    
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
```

```python
# Aqui configuramos os pesos e os bias para visualizar o modelo e a fronteira de decis√£o, posicionando a reta de forma a classificar os dados corretamente

w1 = 5
w2 = 0.8
b = -0.1
plotModel(w1, w2, b)
```

```python
p = (-1, 1)
print(w1*p[0] + w2*p[1] + b)
```

```python
# Aqui descobrimos qual o lado positivo e negativo da reta, para isso, calculamos o valor da fun√ß√£o de decis√£o para um ponto qualquer e verificamos se √© maior ou menor que 0.

def classify(point, weight1, weight2, b):
    ret = weight1*point[0] + weight2*point[1] + b # fun√ß√£o de decis√£o linear == Equa√ß√£o da reta
    if ret > 0:
        return 1, 'yellow'
    else:
        return 0, 'pink'

p = (2, -1)
label, color = classify(p, w1, w2, b)
print(label, color)

plotModel(w1, w2, b)
plt.plot(p[0], p[1], marker='p', color=color, markersize=20)
```

```python
# Verifica a Acur√°cio do modelo

def accuracy(X, Y, w1, w2, b):
    correct = 0
    for i in range(len(X)):
        label, _ = classify(X[i], w1, w2, b)
        if label == Y[i]:
            correct += 1
    return correct / len(X)

print("""Acur√°cia: {:.2f}%""".format(accuracy(X, Y, w1, w2, b)*100))
```
